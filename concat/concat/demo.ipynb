### Writer Class

```python
class Writer:
    def __init__(self):
        self.next_svg_number = 0
        self.image_paths = []
        self.images = []
        
    def write_svgs_to_dir(self, stroke_width=2):
        npy_files = glob.glob('../data/dataset/train/**.json')
        for file in npy_files:
            drawing = json.load(open(file))
            name = get_annotation(file)
            print("rendering", name)
            data, _ = convert_3d(drawing, return_flag=True, threshold=50)
            write_strokes_to_svg(data, stroke_width=stroke_width, crop=True, svg_filepath=f'../svgout/{name}.svg')
        # self.next_svg_number += 1
    
    def write_random_image_to_png_and_get_image_instance(self, stroke_width=2):
        npy_files = glob.glob('../data/dataset/train/**.json')
        json_path = np.random.choice(npy_files)
        drawing = json.load(open(json_path))
        print(get_annotation(json_path))
        data, _ = convert_3d(drawing, return_flag=True, threshold=50)
        image, image_path = draw_strokes(data, stroke_width=stroke_width, crop=True, svg_filepath=f'../svgout/the_strokes{self.next_svg_number}.svg')
        self.next_svg_number += 1
        self.image_paths.append(image_path)
        self.images.append(image)
```

```python
def show_svg(svg_path):
    display(SVG(svg_path))
```

```python
writer = Writer()
```

# Write and View SVG Files

```python
writer.write_svgs_to_dir()
```

```python
for x in range(100):
    writer.write_random_image_to_svg(stroke_width=2)
```
